```python
import os, math, random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import generate_letters
from generate_letters import combine_letters
import utility_functions
```

```python
def generate_letters(list_of_letters):

    img = 255*combine_letters(list_of_letters)
    size = np.shape(img)
    new_size= size/np.max(size)*24
    
    img = cv2.resize(img,dsize = (24,24),interpolation = cv2.INTER_AREA)
    img = utility_functions.add_border(img,2,0)
    
    return img
```

```python
plt.imshow(generate_letters(["Het","Mem","Het"]),cmap ="gray")
```
