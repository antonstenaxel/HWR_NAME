```python
from baseline_cnn_classifier import Classifier
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
cf = Classifier(path_to_model = "Models/baseline_cnn.h5")
```

```python
p = cf.predict("Test_Images/multi_letter.jpg")
```

```python
np.shape(p)
```

```python
dim = np.shape(p)

x = np.arange(0,dim[1])
y = np.arange(0,dim[0])
fig = plt.figure()
ax = fig.gca(projection='3d')

X,Y = np.meshgrid(x,y)
# Plot the surface.
surf = ax.plot_surface(X, Y, p, cmap=cm.coolwarm,linewidth=0, antialiased=False)


# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.xlabel("Sliding window")
plt.ylabel("Letter")

plt.show()
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
fig, ax = plt.subplots()
im = ax.imshow(p,cmap="Blues")

# We want to show all ticks...
ax.set_xticks(np.arange(dim[1]))
ax.set_yticks(np.arange(dim[0]))
# ... and label them with the respective list entries
ax.set_yticklabels([letter for letter in cf.dic.values()])
ax.set_xticklabels(range(dim[1]))

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=0, ha="right", rotation_mode="anchor")

ax.set_title("Probability Map")
fig.tight_layout()
plt.xlabel("Window-step")
#fig.colorbar(im,ax = ax)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)

plt.colorbar(im, cax=cax)
plot = plt.show()


```
