```python
#mat as numpy ndarray
def rescale(mat, scale = [0,1]):
    
    mat -= np.min(mat)
    mat /= np.max(mat)
    
    mat *= (scale[1] - scale[0])
    mat += scale[0]
    
    return mat
```
